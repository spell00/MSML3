```python
# Make plots of stats
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python

path = f'/home/simon/MSML3/results/multi/mz1/rt10/ms2/200spd/thr0.0/all/b14-b13-b12-b11-b10-b9-b8-b7-b6-b5-b4-b3-b2-b1_binary0_-1_gkf0_ovr0_mz0-10000rt0-320_na_h/xgboost'
# Load all csv that starts by test_linear_shap in path and concatenate them, filling missing values with 0
dfs = []
names = []
for file in os.listdir(path):
    if file.startswith('test_linear_shap') and file.endswith('abs.csv'):
        names += [file.split('_')[3]]
        dfs.append(pd.read_csv
                     (os.path.join(path, file), index_col=0).fillna(0))
df = pd.concat(dfs, axis=1)
df = df.fillna(0)
df.columns = names
# Save the concatenated dataframe
df.to_csv(os.path.join(path, 'test_linear_shap_concat.csv'))

# Count the number of times each feature is not 0 per column
counts = df.astype(bool).sum(axis=0)
# Sort the counts
counts = counts.sort_values(ascending=False)
# Plot the histogram of counts
plt.figure(figsize=(6.4, 4.8))
plt.bar(counts.keys(), counts.values)
plt.xticks(rotation=45)
plt.title('Number of times each feature is not 0')
plt.ylabel('#Feature')
plt.xlabel('Class')
plt.show()

# Count the number of times each feature is not 0 per row
counts = df.astype(bool).sum(axis=1)
count_counts = {i: sum(counts == i) for i in range(2, counts.max()+1)}
plt.figure(figsize=(6.4, 4.8))
plt.bar(count_counts.keys(), count_counts.values())
plt.xticks(rotation=45)
plt.title('Number of times each feature is not 0')
plt.xlabel('# Classes')
plt.ylabel('# Features')
plt.show()

# Number of times a features is present across classes
{i: sum(counts == i) for i in range(1, counts.max()+1)}
```
